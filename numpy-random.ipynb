### numpy.random.choice

$$ numpy.random.choice(a, size=None, replace=True, p=None) $$

This function returns a random sample from the non-empty sequence. Index Error occurs if no value (list or sequence) is passed into this function as you cannot make a selection from an empty sequence.  

Generates a random sample from a given 1D array
random.choice(seq)

This function selects a value from a give array, and repeats for the number of times specified. 

```python
fruit= ['apple', 'banana', 'orange', 'pear', 'grapes']

# Generate a random sample from a 1-D array
print("Random choice of fruit: \n", np.random.choice(fruit),"\n") 

# Generate a random sample from a 1-D array
print("Random choice of fruit: \n", np.random.choice(fruit,2),"\n") 

# Generate a random sample from a 1-D array
print("Random choice of fruit: \n", np.random.choice(fruit,(2,3)),"\n") 
 
# ValueError: Cannot take a larger sample than population when 'replace=False'
# print("Random choice of fruit: \n", np.random.choice(fruit,(2,3), replace=False),"\n") 

print("Random choice of fruit: \n", np.random.choice(fruit,(5,2), replace=False),"\n")

```

```python
# Generate a uniform random sample where each fruit has an equal chance of being picked.

fruit = ['apple', 'banana', 'orange', 'pear', 'grapes']
x = np.random.choice(fruit,(5,3))
print(x)
plt.hist(x)
plt.show()

# Generate a non-uniform random sample
# As you can see we have given the probability of each fruit to be selected

fruit= ['apple', 'banana', 'orange', 'pear', 'grapes']
x = np.random.choice(fruit,(5,3), p=[0.4, 0.30, 0.20, 0.10, 0])
print(x)
plt.hist(x)
plt.show()
```

```python
# Adapted from: https://www.science-emergence.com/Articles/How-to-randomly-shuffle-an-array-in-python-using-numpy/

# 1D array using shuffle function, can use strings and numbers:

s = np.array(["Ireland","Italy","Spain","France","America","Wales","Belgium","Scotland"])
np.random.shuffle(s)
print("1D array shuffle:\n ", s, "\n")

'''
Shuffle the 1D array once again, shows that when you shuffle the array again it
does not retuen it in the same order:
'''
np.random.shuffle(s)
print("1D array shuffle again:\n ", s, "\n")

# 2D or more array using the shuffle function:
num = np.array([[2,4,8,16,32],[5,10,15,20,25],[3,6,12,24,48]])
np.random.shuffle(num)
print("2D array shuffle again:\n ", num, "\n")

# Transform 2D to a 1D array:
num = num.ravel()
print("Transform 2D to a 1D array: \n ", num, "\n")

# Reshape array to the original shape: 5 rows and 3 columns
num = num.reshape(5,3)
print("Reshape array to the original shape: \n", num,"\n")
```

### Binomial Distribution:

The probability of exactly x successes in n trials is

$$ P(X = x) = \left(\begin{array}{l}n\\ x\end{array}\right)pxqn − x $$

Properties of a Binomial Distribution are:

1. Each trial is independent of the others.
<i> 
    * A trial is considered independent if the outcome on one trial does not affect the outcome on other trials. Thus the probability of success (p) is the same on every trial. </i>
    
    
$$  p + q = 1  $$ 
<center> <i> where p denotes the probability of success on one trial and q denotes probability of a failure </i></center>


2. There are only two possible outcomes in a trial- either a success or a failure.
<i> Examples:
    * Flip a coin 10 times to see how many heads you get. Note in this instanve head is a success and tail is a failure.
    * Question people to see if they watch the Game of Throne finale - yes (success) and no (failure)
    * Roll a dice 20 times to see the number of times you roll a 6 - 6 (success) and everything else i.e. 1-5 (failure) </i>

3. There are fixed number of trials. Trials can be seen as repetitions of an experiment. 
4. The probability of success and failure is same for all trials. (Trials are identical.)

The random varibale (x) represents the number of successes obtaianed in n independent trials. Mean of x can ve calculated using the formula $ μ = np $ and standard deviation can be calculated using the formula $ σ = \text{ }\sqrt{npq} $

```python
# https://towardsdatascience.com/fun-with-the-binomial-distribution-96a5ecabf65b

trials = 1000       # Number of trials
n = 10              # Number of independent experiments in each trial
p = 0.5             # Probability of success for each experiment

# Coin toss trial function
# heads is a list of the number of successes from each trial of n experiments
def run_binom(trials, n, p):
    heads = []
    for i in range(trials):
        tosses = [np.random.random() for i in range(n)]
        heads.append(len([i for i in tosses if i>=0.50]))
    return heads
# Run the function
heads = run_binom(trials, n, p)

sns.distplot(heads, bins=10, label='simulation results')
plt.xlabel("Number of Heads", fontweight="bold", fontsize="12", color="b")                                         # Label of the x-axis
plt.ylabel("Frequency", fontweight="bold", fontsize="12", color="b")                                                # Label of the y-axis
plt.title("Binomial Distribution", fontweight="bold", fontsize="14", color="r")  
```

### Chi Square Distribution:

There are 2 kinds of chi square tests: test of independence for data and goodness of fit for a model. "Goodness of fit" statistic measures whether there is any difference between the experimental (observed) value and the theoretical (expected) value.  It aims to test how likely that the observed distribution is due to chance.[https://www.ling.upenn.edu › ~clight › chisquared]. 

Chi Square Distribution is used in testing the null hypothesis. 
<b> Properties: </b>
1. This test only works for categorical data e.g. Gender or Color but not numerical data e.g. height or weight.
2. Numbers must be large enough. Each entry must have at least 5.
3. two categorical variables are independent in some population.

Examples of Chi Squared Test:
 
 
$$ numpy.random.chisquare(df, size=None) $$

* <b> df:</b> Number of degrees of freedom.
* <b> size:</b> Output the shape. Default(none) will return single values if df is a scalar. Otherwise, np.array(df).size samples are drawn.
https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.random.chisquare.html

When df independent random variables, each with standard normal distributions (mean 0, variance 1), are squared and summed, the resulting distribution is chi-square (see Notes).

he data used in calculating a chi square statistic must be random, raw, mutually exclusive, drawn from independent variables, and drawn from a large enough sample.

```python

p = np.random.chisquare(2,4)

z = np.random.chisquare(8,4)

sns.distplot(p)
sns.distplot(z)

plt.title('Chisquare Distribution')

plt.show()
```

```python
observed = roscommon_table

# Generate the national political ratio
national_ratios = national_table/len(national) 

# Generate expected counts
expected = national_ratios * len(roscommon)   

# Formula for chi squared stat:
# https://www.statisticshowto.datasciencecentral.com/probability-and-statistics/chi-square/
chi_squared_stat = (((observed-expected)**2)/expected).sum()
print(chi_squared_stat)
```

Note: The chi-squared test assumes none of the expected counts are less than 5.

In the chi-square test we compare the chi-square test statistic to a critical value based on the chi-square distribution. The scipy library shorthand for the chi-square distribution is chi2. 

```python
# critical value for 95% confidence level and check the p-value of our result:

c = stats.chi2.ppf(q = 0.95, df = 4)   
print("Critical value", c)

# Find the p-value
p = 1 - stats.chi2.cdf(x=chi_squared_stat, df=4)
print("P value", p)
```

Test of Independence:

Independence is a key concept in probability that describes a situation where knowing the value of one variable tells you nothing about the value of another.
Here we will test whether 2 categorical variables are independent     

```python
np.random.seed(10)

# Sample data randomly at fixed probabilities
voter_party = np.random.choice(a= ["Fianna Fail", "Fine Gael", "Sinn Fein", "Independent", "Other"], p = [0.05, 0.15 ,0.25, 0.05, 0.5], size=1000)

# Sample data randomly at fixed probabilities
voter_age = np.random.choice(a= ["18 -24 years", "25- 60 years", "Over 60"], p = [0.4, 0.2, 0.4], size=1000)

voters = pd.DataFrame({"party":voter_party, "age":voter_age})

voter_tab = pd.crosstab(voters.party, voters.age, margins = True)

voter_tab.columns = ["18 -24 years","25- 60 years","Over 60", "row_totals"]

voter_tab.index = ["Fianna Fail","Fine Gael","Sinn Fein","Independent","Other","col_totals"]

observed = voter_tab.iloc[0:5,0:3]   # Get table without totals for later use
voter_tab
```

Note that we did not use the race data to inform our generation of the party data so the variables are independent.
We use the same chi squared formula  for Test of independence as that which we used for goodness of fit test However, this time the expected counts for each cell is calculated by multiplying row total x column total for that cell and then divide this by column total number of observations. However, this is very repetitive so there is a much shorter function that can be used np.outer() which will take the total rows and total columns in the table and divide by observations.

```python
expected =  np.outer(voter_tab["row_totals"][0:5], voter_tab.loc["col_totals"][0:3]) / 1000

expected = pd.DataFrame(expected)

expected.columns = ["18 -24 years","25- 60 years","Over 60"]
expected.index = ["Fianna Fail","Fine Gael","Sinn Fein","Independent","Other"]

expected
```

```python
'''
.sum() is called twice i.e. to sum columns and then sum all the column together
 which will thus generate the sum of the table. 

'''
chi_squared_stat = (((observed-expected)**2)/expected).sum().sum()
print(chi_squared_stat)
```

```python
# Find the critical value for 95% confidence*
crit = stats.chi2.ppf(q = 0.95, df = 8)   
print("Critical value", crit)

'''
Note: The degrees of freedom for a test of independence equals the product 
of the number of categories in each variable minus 1. 
In this case we have a 5x3 table so df = 4x2 = 8.
'''

# Find the p-value
p_value = 1 - stats.chi2.cdf(x=chi_squared_stat, df=8)
print("P value", p_value)
```

```python
'''
Expected counts: Due to the high p value the test result does not indicate 
a significant relationship between the variables of political vote and age
The function below is used to conduct a test of independence automatically 
given a frequency table of observed counts.

'''
stats.chi2_contingency(observed= observed)
```
