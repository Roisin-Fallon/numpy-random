<a name="rand"></a>
### rand function:

$$ numpy.random.rand(d0, d1, ..., dn) $$ 

rand creates an array of specified shape and fills it with random values. 
numpy.random.rand generates samples from uniform distribution in the range [0, 1]. [8] Unifrom distribution is described below in detail but esentially it is where all values are equally likely to be selected e.g. in a fair deck of 52 cards each card has the same probability of being selected (i.e. 1/52).

The numbers that you pass into rand are going to determine the dimensions of a returned array. They should all be positive. If we do not pass in any numbers we should get a single python float back. Hence, it returns back an ndarray shape with the numbers that you have passed in. [7]

```python
# Create a 1D array 

'''
Numpy creates arrays and normally it likes to do it so that every row and every column has the 
same number of elements.
This function returns random numbers back in an array of the dimensions that you give it.  
Thus, we can see that the bracket tells us the shape of the array: how many rows and columns the array has. 

'''
# np.random.rand(3) will give an array with just 3 elements in it.

print("1D array with random values: \n", np.random.rand(3), "\n")

# If no value is passed in through the paranthese a single number will be returned.
print("1D array with random values: \n", np.random.rand(), "\n")
```

```python
# Create a 2D array which is an array within an array 
# 3 represents the number of rows and 2 represents the number of columns

''' 
If we re-run the function we would again get an array with 3 rows and 2 columns but the numbers are different. 
Thus, everytime we call this function we may get different numbers in our array back every single time (random numbers generated). 

''' 
print("2D array with random values: \n", np.random.rand(3,2), "\n")
```

```python
# Create a 3D array which is an array within an array within an array 
# this will give you 2 array each of those is a two dimensional array containing 4 rows 3 columns. 

print("3D random values array: \n", np.random.rand(2, 4, 3), "\n")
```

```python
'''
Histogram with 100000 random values generated by rand
As expected it has unifrom distribution which essetially means all values
are equally likely to be selected

'''

r = np.random.rand(100000)

sns.distplot(r, kde=False, color="red")
plt.xlabel("$ x values $", fontweight="bold", fontsize="12", color="b")
plt.ylabel(r'Frequency', fontweight="bold", fontsize="12", color="b")
plt.title("Uniform Distribution using rand function", fontweight="bold", fontsize="14", color="r")
```

$$ numpy.random.randn(d0, d1, ..., dn) $$

Generates an array of a specified shape and fills the shape with random values

### numpy.random.randint

$$ numpy.random.randint(low, high=None, size=None, dtype='l') $$

randint function returns a random integer from low (inclusive) to high (exclusive), both of which must be integer type values.
* Low: refers to the starting point i.e. smallest integer to be taken from the integer. 
* High: refers to the largest number to be drawn from the distribution. This value is optional so if not included the numbers generated will be [0, low].

```python
# Generates a random number from default (0) and up to but not including 3
print("1D random values array: \n", np.random.randint(3), "\n")

# Generates a random number between 0 up to but excluding 20 
print("Random number between 0 and 19 inclusive: \n", np.random.randint(0, 20), "\n")

# 1D array: Generates an array of 10 random number between 0 to 19 inclusive 
print("10 random number between 0 and 19 inclusive: \n", np.random.randint(0, 20, 10), "\n")

# Another method to generate 10 random numbers 20 is the high number
# the default low number is 0
print("1D array of 10 random numbers between 0 and 19 inclusive: \n", np.random.randint(20, size=10),"\n")

# 2D array: Generate a 2 x 3 array of ints between 0 and 19, inclusive:
print("2D array between 0 and 19 inclusive: \n", np.random.randint(0, 20, size=(2, 3)), "\n")

# 3D array: Generate a 2 x 4 array of ints between -12 and 19, inclusive:
print("3D array between -12 and 19 inclusive: \n", np.random.randint(-12, 20, size=(2, 4, 3)))
```

<a name="shuf"></a>
<b><i>numpy.random.shuffle</i></b>

This function will modify a sequence in-place. Essentially this means that the function does not change tht content of an array, it just shuffles it. Hence, only the order of  the sub-array is changed, but never is the content changed. In addition, the shuffle function only shuffles the array onward the first axis in the case of a multi-dimensional array.

```python
# Adapted from: https://www.science-emergence.com/Articles/How-to-randomly-shuffle-an-array-in-python-using-numpy/

# 1D array using shuffle function, can use strings and numbers:

s = np.array(["Ireland","Italy","Spain","France","America","Wales","Belgium","Scotland"])
np.random.shuffle(s)
print("1D array shuffle:\n ", s, "\n")

'''
Shuffle the 1D array once again, shows that when you shuffle the array again it
does not retuen it in the same order:
'''
np.random.shuffle(s)
print("1D array shuffle again:\n ", s, "\n")

# 2D or more array using the shuffle function:
num = np.array([[2,4,8,16,32],[5,10,15,20,25],[3,6,12,24,48]])
np.random.shuffle(num)
print("2D array shuffle again:\n ", num, "\n")

# Transform 2D to a 1D array:
num = num.ravel()
print("Transform 2D to a 1D array: \n ", num, "\n")

# Reshape array to the original shape: 5 rows and 3 columns
num = num.reshape(5,3)
print("Reshape array to the original shape: \n", num,"\n")
```

## Binomial Distribution:

The probability of exactly x successes in n trials is

$$ P(X = x) = \left(\begin{array}{l}n\\ x\end{array}\right)pxqn − x $$

<b><i> Properties of a Binomial Distribution are:</i></b>

1. Each trial is independent of the others.
<i> 
    * A trial is considered independent if the outcome on one trial does not affect the outcome on other trials. Thus the probability of success (p) is the same on every trial. </i>
    
    
$$  p + q = 1  $$ 
<center> <i> where p denotes the probability of success on one trial and q denotes probability of a failure </i></center>


2. There are only two possible outcomes in a trial- either a success or a failure.
<i> Examples:
    * Flip a coin 10 times to see how many heads you get. Note in this instanve head is a success and tail is a failure.
    * Question people to see if they watch the Game of Throne finale - yes (success) and no (failure)
    * Roll a dice 20 times to see the number of times you roll a 6 - 6 (success) and everything else i.e. 1-5 (failure) </i>

3. There are fixed number of trials. Trials can be seen as repetitions of an experiment. 
4. The probability of success and failure is same for all trials. (Trials are identical.)

The random varibale (x) represents the number of successes obtaianed in n independent trials. Mean of x can ve calculated using the formula $ μ = np $ and standard deviation can be calculated using the formula $ σ = \text{ }\sqrt{npq} $  [19]

$$ numpy.random.binomial(n, p, size=None) $$

 -  n denotes the number of trials
 -  p refers to the probability of each trial
 -  size refers to number of times tested 

```python
# result of flipping a coin 20 times, tested 100 times.
np.random.binomial(10, 0.5, 100)
```

```python
# https://towardsdatascience.com/fun-with-the-binomial-distribution-96a5ecabf65b

trials = 1000       # Number of trials
n = 10              # Number of independent experiments in each trial
p = 0.5             # Probability of success for each experiment

# Coin toss trial function
# heads is a list of the number of successes from each trial of n experiments
def run_binom(trials, n, p):
    heads = []
    for i in range(trials):
        tosses = [np.random.random() for i in range(n)]
        heads.append(len([i for i in tosses if i>=0.50]))
    return heads
# Run the function
heads = run_binom(trials, n, p)

sns.distplot(heads, bins=10, label='simulation results')
plt.xlabel("Number of Heads", fontweight="bold", fontsize="12", color="b")                                         # Label of the x-axis
plt.ylabel("Frequency", fontweight="bold", fontsize="12", color="b")                                                # Label of the y-axis
plt.title("Binomial Distribution", fontweight="bold", fontsize="14", color="r")  
```

```python
# # https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.binom.html
# Actual binomial distribution

from scipy.stats import binom

x = binom.rvs(n=10,p=0.8,loc=0,size=10000)  # number of trials, probability of each trial, size
sns.distplot(x,color='red', hist_kws={"linewidth": 20,'alpha':1}) # Generate graph
plt.xlabel("Number of Heads", fontweight="bold", fontsize="12", color="b")                                         # Label of the x-axis
plt.ylabel("Frequency", fontweight="bold", fontsize="12", color="b")                                                # Label of the y-axis
plt.title("Binomial Distribution", fontweight="bold", fontsize="14", color="r") 
```

## Poisson Distribution:

Counts the number of times than any one event is likely to occur within a specified period of time. Deals with independent events which occur ata aconstant rate within given interval of time. 

1. Is a discrete distribution 
    <i> 
      * A discrete function means that an event can only be measured as occuring or not occuring. Thus, the variable measured must be a integer </i>
2. Each occurance is independent of the other occurrences 
3. It describes discrete occurrences over an interval.
4. Occurances in each interval can range from zero to infinity
5. The mean number of occurrences must be constant throughout the experiment. [17]

<b><i>Similarities & differences between Binominal & Poisson Distributions </b></i>

Binomial distribution and Poisson distribution have some simularities but several distinct differences. 

* As stated previously binomial distribution has 2 possible outcomes denoted as a success or failure from a given number of trials. Poisson distribution in comparasion has unlimited number of possible outcomes.
* Binomial Distribution is biparametric (characterised by a two parameter n and p). Poisson distribution is uniparametric (characterised by a single parameter m).
* Binomial has fixed number of trials whereas Poisson had unlimited number of trials.
* Binomial distribution has a constant success probability whereas Poisson distribution had a very small number of success chances. [18]

<b><i>Examples:</i></b>

Estimate how likely it is that something will happen "X" number of times
* Number of car accidents in a year 
* Number of typing errors on a page
* Failure of a machine in one month
* Number of patients that arrive at the ER in a 24hr period


```python
# Intializing the parameters for Poisson distribution
# https://docs.w3cub.com/numpy~1.14/generated/numpy.random.poisson/

s = np.random.poisson(5, 10000) # Draw samples from the parameterized Poisson distribution

count, bins, ignored = plt.hist(s, 14, density=True, facecolor='green', alpha=0.5)    
plt.ylabel("Frequency", fontweight="bold", fontsize="12", color="b")
plt.title("Poisson Distribution", fontweight="bold", fontsize="14", color="r")  
```

```python
# https://docs.w3cub.com/numpy~1.14/generated/numpy.random.poisson/

s = np.random.poisson(lam=(100., 500.), size=(100, 2))
count, bins, ignored = plt.hist(s, 14, density=True, facecolor='green', alpha=0.5)
plt.show()
```

```python
# http://www.learningaboutelectronics.com/Articles/How-to-create-a-poisson-probability-mass-function-plot-in-Python.php

from scipy.stats import poisson
# Creates an x -axis of values that range from 200 to 300 with an increment of 0.5
x= np.arange(200,300,0.5)
# Create a poisson probability mass function with a mean of 250
plt.plot(x, poisson.pmf(x,250))
plt.ylabel("Frequency", fontweight="bold", fontsize="12", color="b") 
plt.title("Probability Density Function", fontweight="bold", fontsize="14", color="r")
```

From the above plot it is clear the probability of 250 being the value selected is high while the probability of 200 and 300 is low in comparasion which could indicate outlier values.

## Normal Distribution:

<b><i> Properties of normal distribution: </b></i>

* It is also known as the Gaussian distribution. Normal distribution occurs often in nature and forms an important part of probability statistics. 

There are 2 main parameters of a normal distribution which are mean and standard deviation. The parameters determine the shape and probability of the distribution. The shape of the normal distribution is a symmetrical bell shaped curve. 

Standard Normal Distribution refers to a curve where the mean is 0 and standard deviation is 1. 

<b><i> Mean </b></i>
* Mean is directly in the middle of the distribution and hence divides the area into two halves
* Mean and the median are the same value due to its symmetrical shape
* Total area under the curve is 1

<b><i> Standard Deviation </b></i>
* Standard deviation measures the dispersion of the data points relative to the mean. It determines how far away from the mean the data points are positioned and represents the distance between the mean and the observations. Standard deviation determines the width of the curve i.e. if the standard deviation is small in relation to mean it creates a steep curve whereas a large standard deviation produces a flat curve. 

<b><i>Empirical Rule: </b></i>



From this image we can denote that 68% of your observations will fall between one standard deviation of the mean. 95% will fall within two, and 99.7% will fall within three.

The function has its peak at the $ \mu  $ , and its “spread” increases with the  $ \sigma $. This implies that numpy.random.normal is more likely to return samples lying close to the mean, rather than those far away.[16]

<b><i> Probability density:</b></i>



* $ \mu  $ is the mean 
* $ \sigma $ is the standard deviation
* $ \sigma^2 $ which is the standard deviation squared, also known as the variance

## Uniform Distribution:

$$ numpy.random.uniform(low=0.0, high=1.0, size=None) $$
                       
- <b>Low:</b><i> Refers to lower bound. Values Generated will be greater than or equal to low.  Default is 0. </i>             
- <b>High:</b><i> Refers to upper bound. Values Generated will be less than high. Default is 1. </i>
- <b>Size:</b><i> Output the shape. Default(none) will return single values.</i>

<b><i> Properties: </i></b>

* Uniform distributions is symmetric.
* These are probability distributions where any value within the given interval is equally likely to be drawn. Thus, the distribution is basically a straight line. 
* In uniform distribution the mean and median coincide. 
* 2 types of uniform distributions: discrete and continuous. 
    * Outcome of discrete distributions are discrete
    * Outcomes of continuous distribution are continuous and infinite.
    
<b><i> Example:</i></b>
1. Roll a fair dice the probability of getting the number 2 (or any other number from 1 to 6) is 1/6.
2. Picking a queen of heart (or any other card) in a fair deck its probability is 1/52.

```python
# Descriptve statistics to see if the mean an medians coincide

un = np.random.uniform(-1,0,1000)
print("Mean = ", np.mean(un))
print("Median = ",np.median(un))
```

## Chi Square Distribution:

The chi squared distribution is a continuous probability distribution that is widely used in statistical inference. There are 2 kinds of chi square tests: test of independence for data and goodness of fit for a model. "Goodness of fit" statistic measures whether there is any difference between the experimental (observed) value and the theoretical (expected) value.  It aims to test how likely that the observed distribution is due to chance.[20]. Chi Square Distribution is used in testing the null hypothesis. 

<b><i> Properties: </i></b>
1. This test only works for categorical data e.g. Gender or Color but not numerical data e.g. height or weight.
2. Numbers must be large enough. Each entry must have at least 5.
3. two categorical variables are independent in some population.

<b><i>Examples of Chi Squared Test:</i></b>
 
 
$$ numpy.random.chisquare(df, size=None) $$

* <b> df:</b> Number of degrees of freedom.
* <b> size:</b> Output the shape. Default(none) will return single values if df is a scalar. Otherwise, np.array(df).size samples are drawn.[21]

When df independent random variables, each with standard normal distributions (mean 0, variance 1), are squared and summed, the resulting distribution is chi-square (see Notes).

he data used in calculating a chi square statistic must be random, raw, mutually exclusive, drawn from independent variables, and drawn from a large enough sample.

```python
#  This shows the different degrees of freedom (k values)
# Adapted from: https://www.astroml.org/book_figures/chapter3/fig_chi2_distribution.html

from scipy.stats import chi2

# Define the distribution parameters to be plotted

x = np.linspace(0, 10, 100)
k_values = [1, 2, 6, 8]
mu = 0

# plot all the line styles on one distribution plot
linestyles = ['-', '--', ':', '-.']
fig.subplots_adjust(bottom=0.12)

for k, ls in zip(k_values, linestyles):
    dist = chi2(k, mu)
    plt.plot(x, dist.pdf(x), ls=ls, c='black', label=r'$k=%i$' % k)

# Limits on the axis
plt.xlim(0, 10)
plt.ylim(0, 0.5)

# Formatting the plot
plt.xlabel("Value", fontweight="bold", fontsize="12", color="b")
plt.ylabel("Frequency", fontweight="bold", fontsize="12", color="b")
plt.title("Chi Square Distribution", fontweight="bold", fontsize="14", color="r") 
plt.legend()
plt.show()
```

```python
observed = roscommon_table

# Generate the national political ratio
national_ratios = national_table/len(national) 

# Generate expected counts
expected = national_ratios * len(roscommon)   

# Formula for chi squared stat:
# https://www.statisticshowto.datasciencecentral.com/probability-and-statistics/chi-square/
chi_squared_stat = (((observed-expected)**2)/expected).sum()
print(chi_squared_stat)
```

Note: The chi-squared test assumes none of the expected counts are less than 5.

In the chi-square test we compare the chi-square test statistic to a critical value based on the chi-square distribution. The scipy library shorthand for the chi-square distribution is chi2. 

```python
# critical value for 95% confidence level and check the p-value of our result:

c = stats.chi2.ppf(q = 0.95, df = 4)   
print("Critical value", c)

# Find the p-value
p = 1 - stats.chi2.cdf(x=chi_squared_stat, df=4)
print("P value", p)
```

<b><i>Test of Independence:</i></b>

Independence is a key concept in probability that describes a situation where knowing the value of one variable tells you nothing about the value of another.
Here we will test whether 2 categorical variables are independent     

```python
np.random.seed(10)

# Sample data randomly at fixed probabilities
voter_party = np.random.choice(a= ["Fianna Fail", "Fine Gael", "Sinn Fein", "Independent", "Other"], p = [0.05, 0.15 ,0.25, 0.05, 0.5], size=1000)

# Sample data randomly at fixed probabilities
voter_age = np.random.choice(a= ["18 -24 years", "25- 60 years", "Over 60"], p = [0.4, 0.2, 0.4], size=1000)

voters = pd.DataFrame({"party":voter_party, "age":voter_age})

voter_tab = pd.crosstab(voters.party, voters.age, margins = True)

voter_tab.columns = ["18 -24 years","25- 60 years","Over 60", "row_totals"]

voter_tab.index = ["Fianna Fail","Fine Gael","Sinn Fein","Independent","Other","col_totals"]

observed = voter_tab.iloc[0:5,0:3]   # Get table without totals for later use
voter_tab
```

Note that we did not use the race data to inform our generation of the party data so the variables are independent.
We use the same chi squared formula  for Test of independence as that which we used for goodness of fit test However, this time the expected counts for each cell is calculated by multiplying row total x column total for that cell and then divide this by column total number of observations. However, this is very repetitive so there is a much shorter function that can be used np.outer() which will take the total rows and total columns in the table and divide by observations.

```python
expected =  np.outer(voter_tab["row_totals"][0:5], voter_tab.loc["col_totals"][0:3]) / 1000

expected = pd.DataFrame(expected)

expected.columns = ["18 -24 years","25- 60 years","Over 60"]
expected.index = ["Fianna Fail","Fine Gael","Sinn Fein","Independent","Other"]

expected
```

```python
'''
.sum() is called twice i.e. to sum columns and then sum all the column together
 which will thus generate the sum of the table. 

'''
chi_squared_stat = (((observed-expected)**2)/expected).sum().sum()
print(chi_squared_stat)
```

```python
# Find the critical value for 95% confidence*
crit = stats.chi2.ppf(q = 0.95, df = 8)   
print("Critical value", crit)

'''
Note: The degrees of freedom for a test of independence equals the product 
of the number of categories in each variable minus 1. 
In this case we have a 5x3 table so df = 4x2 = 8.
'''

# Find the p-value
p_value = 1 - stats.chi2.cdf(x=chi_squared_stat, df=8)
print("P value", p_value)
```

```python
'''
Expected counts: Due to the high p value the test result does not indicate 
a significant relationship between the variables of political vote and age
The function below is used to conduct a test of independence automatically 
given a frequency table of observed counts.

'''
stats.chi2_contingency(observed= observed)
```

## Use of seeds in generating pseudorandom numbers

$$ numpy.random.RandomState $$

Raandom number is a number that is generated from a infinite set or finite range, in which each number in the range is eqaully likely to be selected. True random numbers are only generated by observations of random physical events e.g flip coin or roll dice. [10] 

It is important to understand that the majority of random data generated with computer programs is not fully random in the scientific sense but pseudorandom. Pseudorandm number generator (PRNG) is some type of determinitic algorithm for generating random but still reproducible data. Thus pseudorandom numbers are depend on deterministic algorithms and a seed value (starting value). The reason pseudorandom are used instead of random numbers is fast and secure.

<b> Seed: </b>
The genertation of a random number requires a seed (starting) value. The example below shows that if the same seed value is used, the same sequence of numbers is guaranteed.  

Generally, you want to seed your random number generator with some value that will change each execution of the program. For instance, the current time is a frequently-used seed. The reason why this doesn't happen automatically is so that if you want, you can provide a specific seed to get a known sequence of numbers. [11]

```python
# random module is imported 

import random  
```

```python
'''
$$ numpy.random.seed $$

This method is called when RandomState is initialized.

If you set the np.random.seed(a_fixed_number) every time you call the numpy's
other random function, the result will be the same:
'''

# np.random.seed(0) makes the random numbers predictable
np.random.seed(0)
print("Random number generated using seeed: \n",np.random.rand(3), "\n")

# np.random.seed(0) makes the random numbers predictable
np.random.seed(0)
print("Random number generated using seeed: \n", np.random.rand(3))
```

<b><i><u> Use of random numbers: </u></i></b>

Cryography, games, GPS and many statistical models use random numbers
Crytography are keys for encryption of data. Encryption keys are an important part of computer security as it protects data from unauthorized access over the internet.

```python
'''
Using the random.seed() function of a random module it is possible 
that the shuffling produces the same result every time.

'''

# How to use a random.seed function with the random.shuffle function
# Adapted frm: https://pynative.com/python-random-shuffle/

numbers = [25, 50, 75, 100, 125, 150]
print ("Original list: ", numbers)

# As you can see by using seed(2) it gives the starting position 
random.seed(2)
random.shuffle(numbers)
print("Reshuffled list ", numbers)

numbers = [25, 50, 75, 100, 125, 150]
random.seed(2)
random.shuffle(numbers)
print("Reshuffled list ", numbers)
```
